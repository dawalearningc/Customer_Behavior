```python
# we have the dataset in customer_dataset.csv
df = pd.read_csv('customer_dataset.csv')
df.head()
```

```python
# we use median to impute because it is outlier safe
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns= df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.columns
```

```python
df=df.rename(columns={
    'purchase_amount_(usd)':'purchase_amount'
})
df.columns
```

```python
# creating a new column age_group
labels=['Young Adult','Adult','Middle-aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels) # pd.qcut cuts the data into quantiles
```

```python
df[['age','age_group']].head()
```

```python
# create a new column purchase_frequency_days
frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Bi-weekly':14,
    'Annually':365,
    'Every 3 Months':90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```
